```python
# Tokenize the words

import nltk
from nltk.corpus import stopwords
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
# word_tokens = word_tokenize(example_sent)

df_tweets = df["Tweets"]
word_token_list = []
filtered_tweet = []

for i in range(len(df_tweets.index)):
#     print(df_tweets[i])
    word_token = word_tokenize(df_tweets[i])
    word_token_list.append(word_token)
    
for j in word_token_list:
#     print(j)
    for k in j:
        if k not in stop_words:
            filtered_tweet.append(k)

            
            
filtered_tweet
# list(filter(lambda word: word not in stop_words, df_tweets[0]))
# df_tweets["Tweets"].apply(lambda x: [item for item in x if item not in stop_words])
```

```python
# adding the tokenized list to a dataframe

df1 = pd.DataFrame(word_token_list)
df1['tokenized_tweets']=df1.apply(lambda row: row.dropna().tolist(), axis=1)
df_token = df1[['tokenized_tweets']]
df_token
```

```python
# new df to combine tokenized df with original df

df_nlp = pd.concat([df.reset_index(drop=True), df_token], axis=1)
df_nlp
```

```python
# removing stowords from tweets and adding it to the dataframe

from nltk.corpus import stopwords
stop = stopwords.words('english')


df_nlp['tweets_without_stopwords'] = df_nlp['Tweets'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_nlp
```
